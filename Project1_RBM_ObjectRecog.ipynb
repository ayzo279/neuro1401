```python
from PIL import Image, ImageDraw
import math
import os
import random

def create_square_image(center, square_size, angle, folder, filename):
    size = (256, 256)  # Image size
    image = Image.new("L", size, "white")  # Create a new white image
    draw = ImageDraw.Draw(image)  # Create a draw object

    # Calculate the coordinates of the square's corners before rotation
    half_size = square_size / 2
    corners = [
        (center[0] - half_size, center[1] - half_size),
        (center[0] + half_size, center[1] - half_size),
        (center[0] + half_size, center[1] + half_size),
        (center[0] - half_size, center[1] + half_size),
    ]

    # Rotate the square
    theta = math.radians(angle)  # Convert angle to radians
    rotated_corners = []
    for x, y in corners:
        # Translate point to origin
        temp_x = x - center[0]
        temp_y = y - center[1]
        # Rotate point
        rotated_x = temp_x * math.cos(theta) - temp_y * math.sin(theta)
        rotated_y = temp_x * math.sin(theta) + temp_y * math.cos(theta)
        # Translate point back
        rotated_x += center[0]
        rotated_y += center[1]
        rotated_corners.append((rotated_x, rotated_y))

    # Draw the rotated square
    draw.polygon(rotated_corners, outline="black")

    # Ensure the folder exists
    if not os.path.exists(folder):
        os.makedirs(folder)

    # Save the image
    image_path = os.path.join(folder, filename)
    image.save(image_path)

# Parameters
n_images = 30
folder_name = "Test"

for i in range(n_images):
    center = (random.randint(256/2 - 40, 256/2+40), random.randint(256/2 - 40, 256/2+40))
    square_size = random.randint(64, 128)
    angle = random.randint(0, 360)
    filename = f"square_{i}.png"
    create_square_image(center, square_size, angle, folder_name, filename)

folder_name = "Training"

for i in range(30):
    center = (random.randint(256/2 - 40, 256/2+40), random.randint(256/2 - 40, 256/2+40))
    square_size = random.randint(64, 128)
    angle = random.randint(0, 360)
    filename = f"square_{i}.png"
    create_square_image(center, square_size, angle, folder_name, filename)


folder_name = "SquareTraining"

for i in range(n_images):
    center = (random.randint(256/2 - 40, 256/2+40), random.randint(256/2 - 40, 256/2+40))
    square_size = random.randint(64, 128)
    angle = random.randint(0, 45)
    filename = f"square_{i}.png"
    create_square_image(center, square_size, angle, folder_name, filename)

folder_name = "SquareTest2"
for i in range(n_images):
    center = (random.randint(256/2 - 40, 256/2+40), random.randint(256/2 - 40, 256/2+40))
    square_size = random.randint(64, 128)
    angle = random.randint(0, 45)
    filename = f"square_{i}.png"
    create_square_image(center, square_size, angle, folder_name, filename)

folder_name = "SquareTest"

for i in range(n_images):
    center = (random.randint(256/2 - 40, 256/2+40), random.randint(256/2 - 40, 256/2+40))
    square_size = random.randint(64, 128)
    angle = random.randint(46, 89)
    filename = f"square_{i}.png"
    create_square_image(center, square_size, angle, folder_name, filename)


def create_triangle_image(center, size, angle, folder, filename):
    image_size = (256, 256)  # Image size
    image = Image.new("L", image_size, "white")  # Create a new white image
    draw = ImageDraw.Draw(image)  # Create a draw object

    # Calculate the triangle's vertices before rotation
    # Assuming an equilateral triangle for simplicity
    height = size * math.sqrt(3) / 2
    vertices = [
        (center[0], center[1] - 2/3 * height),  # Top vertex
        (center[0] - size/2, center[1] + 1/3 * height),  # Bottom left vertex
        (center[0] + size/2, center[1] + 1/3 * height),  # Bottom right vertex
    ]

    # Rotate the triangle
    theta = math.radians(angle)  # Angle in radians
    rotated_vertices = []
    for x, y in vertices:
        # Translate point to origin (center), rotate, translate back
        temp_x = x - center[0]
        temp_y = y - center[1]
        rotated_x = temp_x * math.cos(theta) - temp_y * math.sin(theta) + center[0]
        rotated_y = temp_x * math.sin(theta) + temp_y * math.cos(theta) + center[1]
        rotated_vertices.append((rotated_x, rotated_y))

    # Draw the rotated triangle
    draw.polygon(rotated_vertices, outline="black")

    # Ensure the folder exists
    if not os.path.exists(folder):
        os.makedirs(folder)

    # Save the image
    image_path = os.path.join(folder, filename)
    image.save(image_path)


def create_rectangle_image(center, width, height, angle, folder, filename):
    image_size = (256, 256)  # Image size
    image = Image.new("L", image_size, "white")  # Create a new white image
    draw = ImageDraw.Draw(image)  # Create a draw object

    # Calculate the rectangle's vertices before rotation
    half_width = width / 2
    half_height = height / 2
    vertices = [
        (center[0] - half_width, center[1] - half_height),  # Top left vertex
        (center[0] + half_width, center[1] - half_height),  # Top right vertex
        (center[0] + half_width, center[1] + half_height),  # Bottom right vertex
        (center[0] - half_width, center[1] + half_height),  # Bottom left vertex
    ]

    # Rotate the rectangle
    theta = math.radians(angle)  # Angle in radians
    rotated_vertices = []
    for x, y in vertices:
        # Translate point to origin (center), rotate, translate back
        temp_x = x - center[0]
        temp_y = y - center[1]
        rotated_x = temp_x * math.cos(theta) - temp_y * math.sin(theta) + center[0]
        rotated_y = temp_x * math.sin(theta) + temp_y * math.cos(theta) + center[1]
        rotated_vertices.append((rotated_x, rotated_y))

    # Draw the rotated rectangle
    draw.polygon(rotated_vertices, outline="black")

    # Ensure the folder exists
    if not os.path.exists(folder):
        os.makedirs(folder)

    # Save the image
    image_path = os.path.join(folder, filename)
    image.save(image_path)

# Example usage
folder_name = "Training"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Generate 30 triangles with varying sizes and orientations
for i in range(30):
    center = (random.randint(128 - 40, 128+40),random.randint(128 - 40, 128+40))  # Center of the image
    size = random.randint(64, 128)  # Random size
    angle = random.randint(0, 360)  # Random orientation
    filename = f"triangle_{i}.png"
    create_triangle_image(center, size, angle, folder_name, filename)

# Generate 30 rectangles with varying sizes, orientations
for i in range(30):
    center = (random.randint(128 - 40, 128 + 40), random.randint(128 - 40, 128 + 40))  # Center of the image
    width = random.randint(64, 128)  # Random width
    height = random.randint(64, 128)  # Random height
    while abs(width - height) < 10:
        width = random.randint(64, 128)  # Random width
        height = random.randint(64, 128)  # Random height
    angle = random.randint(0, 360)  # Random orientation
    filename = f"rectangle_{i}.png"
    create_rectangle_image(center, width, height, angle, folder_name, filename)

# Example usage
folder_name = "Test"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Generate 30 triangles with varying sizes and orientations
for i in range(30):
    center = (random.randint(128 - 40, 128+40),random.randint(128 - 40, 128+40))  # Center of the image
    size = random.randint(64, 128)  # Random size
    angle = random.randint(0, 360)  # Random orientation
    filename = f"triangle_{i}.png"
    create_triangle_image(center, size, angle, folder_name, filename)

# Generate 30 rectangles with varying sizes, orientations
for i in range(30):
    center = (random.randint(128 - 40, 128 + 40), random.randint(128 - 40, 128 + 40))  # Center of the image
    width = random.randint(64, 128)  # Random width
    height = random.randint(64, 128)  # Random height
    while abs(width - height) < 10:
        width = random.randint(64, 128)  # Random width
        height = random.randint(64, 128)  # Random height
    angle = random.randint(0, 360)  # Random orientation
    filename = f"rectangle_{i}.png"
    create_rectangle_image(center, width, height, angle, folder_name, filename)
```

## Building the Network


```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from PIL import Image
import os
import random
import matplotlib.pyplot as plt

class RBFNetwork:
  """
  Parameters:
    X: training set of image vectors
    y: training labels corresponding to X
    num_views:
      Number of views, or total number of images presented
    sigma:
      RBF width, larger sigma provides greater flexibility in determining identity of image
    lamb:
      lambda, regularization parameter
    num_centers:
      Number of centers. In training, centers are prototypical image vectors that other input image vectors
      are trained to cluster around. We weigh each input around the label of their nearest centers.
      We choose centers randomly for now, but there are ways to select optimal centers methodically.
  """
  def __init__(self, X, y, num_views, sigma, lamb, num_centers=0):
    self.num_views = num_views
    if num_centers == 0:
      self.num_centers = num_views
    else:
      self.num_centers = num_centers
    self.sigma = sigma
    self.lamb = lamb
    self.weights = np.zeros(len(y))
    self.G = np.zeros((self.num_views, self.num_centers))
    self.X = X
    self.y = y

  """
    Compute the gaussian radial basis function output of a particular point X[i]
  """
  def gaussianRBF(self, X, i, j):
    res = np.exp(-np.linalg.norm(X[i] - X[j])/self.sigma)
    return res

  """
    Randomly sample centers from total images
  """
  def generateCenters(self):
    random.seed(1401) # Fix random number generator output for consistency across tests
    self.centers = random.sample(range(len(self.y)), self.num_centers)

  """
    build the Gmatrix of activations using gaussianRBF computations of each input unit
  """
  def constructGMatrix(self, X):
    G = np.zeros((self.num_views, self.num_views))
    for i in range(self.num_views):
      for j in self.centers:
        G[i][j] = self.gaussianRBF(X, i, j)
    return G

  """
    Compute the weight of each RBF unit
  """
  def computeWeight(self):
    I = np.eye(self.G.shape[1])
    return np.linalg.pinv(self.G.T @ self.G + self.lamb * I) @ self.G.T @ self.y

  """
    Perform model fiting using Xtrain and ytrain
  """
  def fit(self):
    self.generateCenters()
    self.G = self.constructGMatrix(self.X)
    self.weights = self.computeWeight()
    return self

  """
    Predicts the labels of a test set of images Xtest
  """
  def predict(self, Xtest):
    Gtest = self.constructGMatrix(Xtest)
    return Gtest @ self.weights

  """
    Score the accuracy of the model out of 1.0
  """
  def score(self, Xtest, ytest):
    ypred = self.predict(Xtest)
    correct = 0
    for i in range(len(ypred)):
      if int(ypred[i]) == ytest[i]:
        correct += 1
    return correct/len(ytest)


```

```python
def loadImages(folder, size):
  images = []
  labels = []  # Assuming subfolder names are labels
  for label in os.listdir(folder):
    img_path = os.path.join(folder, label)
    with Image.open(img_path) as img:
        img = img.resize(size).convert('L')  # Convert to grayscale
        images.append(np.array(img).flatten())  # Flatten image
        if 'square' in img_path:
          labels.append(1)
        elif 'triangle' in img_path:
          labels.append(0)
        else:
          labels.append(2)
  return np.array(images), np.array(labels)

image_folder = 'Training/'
image_size = (258, 258)  # Resize images to a uniform size
image_folder2 = 'Test/'

# Load images from generated sets
Xtrain, ytrain = loadImages(image_folder, image_size)
Xtest, ytest = loadImages(image_folder2, image_size)

# Normalize input vectors
scaler = StandardScaler()
Xtrain = scaler.fit_transform(Xtrain)
Xtest = scaler.fit_transform(Xtest)

# Create RBF Model and fit data
# random.seed(1401) # Fix random number generator output for consistency across tests
# rbf_model = RBFNetwork(Xtrain, ytrain, len(ytrain), 40, 1, 40)
# rbf_model.fit()

# Score accuracy
# print(rbf_model.score(Xtest, ytest))

lambdas = list(range(-9, 1, 1))
num_centers = list(range(1, 11, 1))
sigmas = list(range(10,101,10))
lambda_output = []
sigma_output = []
centers_output = []

for sigma in sigmas:
  rbf_model = RBFNetwork(Xtrain, ytrain, len(ytrain), sigma = sigma, lamb = 0.01, num_centers = int(0.3 * len(ytrain)))
  rbf_model.fit()
  sigma_output.append(rbf_model.score(Xtest, ytest))

plt.plot(sigmas, sigma_output, color="forestgreen")
plt.title("Width Effect on RBF Model Accuracy")
plt.xlabel(r"$\sigma$")
plt.ylabel("Accuracy")
plt.tight_layout()
plt.show()

# Verify that weights decrease with increase in lambda
# for lamb in lambdas:
#   rbf_model = RBFNetwork(Xtrain, ytrain, len(ytrain), 50, 10 ** lamb, int(0.8 * len(ytrain)))
#   rbf_model.fit()
#   lambda_output.append(np.sum(rbf_model.weights))

# plt.plot(lambdas, lambda_output, color="forestgreen")
# plt.title("Regularization Effect on RBF Model Accuracy")
# plt.xlabel(r"$\log_{10} \lambda$")
# plt.ylabel("Accuracy")
# plt.tight_layout()
# plt.show()

lambda_output = []
for lamb in lambdas:
  rbf_model = RBFNetwork(Xtrain, ytrain, len(ytrain), 50, 10 ** lamb, int(0.8 * len(ytrain)))
  rbf_model.fit()
  lambda_output.append(rbf_model.score(Xtest, ytest))

plt.plot(lambdas, lambda_output, color="forestgreen")
plt.title("Regularization Effect on RBF Model Accuracy")
plt.xlabel(r"$\log_{10} \lambda$")
plt.ylabel("Accuracy")
plt.tight_layout()
plt.show()


for ncenter in num_centers:
  rbf_model = RBFNetwork(Xtrain, ytrain, len(ytrain), 50, 0.01, int(ncenter * len(ytrain)/10))
  rbf_model.fit()
  centers_output.append(rbf_model.score(Xtest, ytest))

plt.plot(num_centers, centers_output, color="forestgreen")
plt.title("Stored Views Effect on RBF Model Accuracy")
plt.xlabel(r"$prop. centers$")
plt.ylabel("Accuracy")
plt.tight_layout()
plt.show()


mat = np.zeros((10,10))
for i, lamb in enumerate(lambdas):
    for j, ncenter in enumerate(num_centers):
      rbf_model = RBFNetwork(Xtrain, ytrain, len(ytrain), 50, 10**lamb, int(ncenter * len(ytrain)/10))
      rbf_model.fit()
      mat[j][i] = rbf_model.score(Xtest, ytest)

matplot = plt.matshow(mat, cmap="RdYlGn")
plt.xlabel(r"$\log_{10} \lambda$")
plt.ylabel(r"$Proportion of center views$")
plt.title("Accuracy")
plt.colorbar(matplot)
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')  # Set tick position
ax.xaxis.set_label_position('bottom')  # Set label position
ax.invert_yaxis()
plt.xticks(np.arange(0, 10, 1), lambdas)
plt.yticks(np.arange(0, 10, 1), num_centers)
plt.show()

mat = np.zeros((10,10))
for i, lamb in enumerate(lambdas):
    for j, sigma in enumerate(sigmas):
      rbf_model = RBFNetwork(Xtrain, ytrain, len(ytrain), sigma, 10**lamb, int(0.8*len(ytrain)))
      rbf_model.fit()
      mat[j][i] = rbf_model.score(Xtest, ytest)

matplot = plt.matshow(mat, cmap="RdYlGn")
plt.xlabel(r"$\log_{10} \lambda$")
plt.ylabel(r"$\sigma$")
plt.title("Accuracy")
plt.colorbar(matplot)
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')  # Set tick position
ax.xaxis.set_label_position('bottom')  # Set label position
ax.invert_yaxis()
plt.xticks(np.arange(0, 10, 1), lambdas)
plt.yticks(np.arange(0, 10, 1), sigmas)
plt.show()

# Predictions:
## Bigger sigma should reduce accuracy as clusters become too big and the model confuses
### squares with rectangles for example.
## Smaller lambda improves accuracy by reducing overfitting?
## Bigger num_centers improves accuracy to a certain extent to account for orientation complexities
### but too many centers will reduce accuracy due to overfitting
```
